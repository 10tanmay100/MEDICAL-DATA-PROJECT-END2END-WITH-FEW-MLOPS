```python
import boto3
import zipfile
import io
import os

def download_files_from_s3(bucket_name, prefix):
    """
    Download files from S3 bucket that match the given prefix.

    :param bucket_name: Name of the S3 bucket.
    :param prefix: Prefix to filter the files in the bucket.
    :return: List of downloaded files.
    """
    s3 = boto3.client('s3')
    objects = s3.list_objects_v2(Bucket=bucket_name, Prefix="data")['Contents']
    files = []
    for obj in objects:
        file_name = obj['Key']
        print(file_name)
        if file_name.endswith('.csv'):
            print("hotcha")
            path=os.path.join("data",file_name.split("/")[1])
            s3.download_file(bucket_name, file_name, path)
            files.append(path)
    return files


def create_zip_in_memory(files):
    """
    Create a ZIP file in memory from a list of files, optionally including an additional file.

    :param files: List of file paths to be zipped.
    :param additional_file: Path to an additional file to include in the ZIP.
    :return: BytesIO object representing the ZIP file.
    """
    zip_buffer = io.BytesIO()
    with zipfile.ZipFile(zip_buffer, 'a', zipfile.ZIP_DEFLATED, False) as zip_file:
        for file in files:
            zip_file.write(file)
    zip_buffer.seek(0)
    return zip_buffer

def upload_zip_to_s3(zip_buffer, bucket_name, zip_name):
    """
    Upload a ZIP file from a BytesIO object to S3.

    :param zip_buffer: BytesIO object containing the ZIP file.
    :param bucket_name: Name of the S3 bucket to upload to.
    :param zip_name: Name for the ZIP file in S3.
    """
    s3 = boto3.resource('s3')
    bucket = s3.Bucket(bucket_name)
    bucket.upload_fileobj(zip_buffer, zip_name)

# Usage example
bucket_name = 'mlflow-test-own'
prefix = 'data.zip'  # Adjust as needed
zip_name = 'output.zip'  # Name of the output ZIP file
additional_csv = 'train1.csv'  # Path to your additional CSV file

csv_files = download_files_from_s3(bucket_name, prefix)
zip_buffer = create_zip_in_memory(csv_files)
upload_zip_to_s3(zip_buffer, bucket_name, zip_name)

# # Clean up local CSV files (optional)
# for file in csv_files + [additional_csv]:
#     os.remove(file)
```
